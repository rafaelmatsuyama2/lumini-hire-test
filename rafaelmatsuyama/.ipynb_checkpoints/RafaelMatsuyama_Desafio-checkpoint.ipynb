```python
import pandas as pd

# Carregando a massa de dados direto como um dataframe.
df = pd.read_csv("desafio_fraude.csv")

# Dando um bizu nos dados para ver se estão ok.
df.head()

# Insight: A coluna Ocorrencia está apresentando diversos valores repetidos, será que a informação seria um ID de cliente ou do cartão de crédito? (Hipótese)

# Insight: A coluna Sacado aparentemente é do valor da transação (Hipótese).

# Insight: As colunas PP1-PP28 todas aparentemente contém valores quantitativos e na mesma ordem de grandeza, algumas colunas não estão normalizadas em relação a uma probabilidade 1.

# Insight: As colunas PP1-PP28 aparentemente são de indicadores normalizados.
```

```python
# Analisando a dimensão da massa de dados.
print(df.shape)

# Checando se todos os registros estão rotulados corretamente (com zero ou um).
print(df[df['Fraude'] == 0].shape)
print(df[df['Fraude'] == 1].shape)

# Insight: A massa de dados está bem desbalanceada, existem poucos positivos para fraude em relação aos negativos, em torno de 0,16% das amostras.

# Analisando as situações que foram rotuladas como fraude.
df[df['Fraude'] == 1]

# Insight: Existem alguns valores estranhos na coluna Sacado para as situações de Fraude, como -1.00 e -0.00 aparecendo com uma elevada frequência. Será que esses dados estão corretos?
# Será que essas transações são "testes" para o fraudador verificar se um cartão está válido antes de conduzir mais transações?
```

```python
# Verificando se essas situações de valores -1.0 aparecem em transações rotuladas como não sendo fraude.
df[df['Sacado'] == -1.00].shape

# Insight: Em torno de 8000 registros com valores de 1 real? Estranho isso. Parece ser validação de meio de pagamento. (Hipótese)
```

```python
# Insight: Tirando esse ponto identificado na coluna Sacado, aparentemente a massa de dados está bem comportada.

# Pretendo utilizar um classificador simples (Logistic Regression) nesta situação, pelos seguintes fatores:
# 1-) É um classificador quase linear, então se a massa de dados conter ligeiras não-linearidades nos parâmetros, consigo capturar isso.
# 2-) Consigo obter os pesos associados a cada parâmetro, para medir o impacto dele no resultado.
# 3-) Occam's Principle, se esse classificador simples obter uma acurácia razoável, a questão passa a ser avaliar o ganho marginal de se utilizar técnicas mais complexas.
```

```python
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Separação das massas de dados de treino e de teste (para validação).
# Usei a divisão clássica de 80:20 (80% para treinamento e 20% para teste).
train, test = train_test_split(df, test_size=0.2)

# Separando o output com os rótulos em variável a parte.
y = train.iloc[:,30]

# Optei por excluir a coluna Ocorrencia para evitar que esta coluna se torne um fator a ser avaliado quantitativamente no modelo. (Hipótese)
X = train.iloc[:,1:30]

# Inicialmente utilizei o solver lbfgs que é o default dele, mas não estava convergindo mesmo com valores elevados de iterações. (Para análise posterior)
# Migrei o solver para o liblinear, pois como a massa de dados é relativamente pequena, não terá muita penalidade em termos computacionais para sua execução.
LR = LogisticRegression(random_state=0, solver='liblinear', multi_class='ovr', max_iter=100).fit(X, y)

# Output rotulado da massa de testes.
yt = test.iloc[:,30]

# Inputs da massa de testes.
Xt = test.iloc[:,1:30]

# Medição da acurácia com 5 casas decimais.
print(round(LR.score(Xt,yt), 5))

# Foi possível obter um modelo com uma razoável acurácia, em torno de 99,9%.
# O valor pode variar um pouco em função do shuffle utilizado no método train_test_split().

# Como a acurácia foi razoável, vou optar por continuar a análise pelos fatores que levaram a essa acurácia, ao invés de analisar outras técnicas (como Decision Trees, Neural Networks ou até mesmo um SVM/PCA).
```

```python
# Listando os pesos para cada parâmetro, arrendondando para duas casas decimais de precisão e colocando em um dataframe.
weights = pd.DataFrame(LR.coef_.round(2))

# Colocando os nomes das colunas para ficar mais prática a visualização.
weights.columns = df.columns.values.tolist()[1:-1]

# Aumentei o threshold de colunas a ser exibida para poder observar todos os pesos na tela.
pd.set_option('display.max_columns', 50)

weights
# Observando os pesos, é possível identificar correlações positivas fortes em PP10 e PP14, seguida por PP12 e PP13.

# Existe uma correlação negativa razoável em PP4, seguida por PP11 e PP21.

# As demais variáveis apresentam influência menor neste modelo preditivo.
# Como os intervalos dentro dos parâmetros são semelhantes, no caso de uma simplificação do modelo, podem ser descartadas sem incorrer em uma redução significativa na acurácia.
```
