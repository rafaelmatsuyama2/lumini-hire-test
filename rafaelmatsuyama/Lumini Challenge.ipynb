```python
import pandas as pd

# Carregando a massa de dados direto como um dataframe.
df = pd.read_csv("Microdados_Enem_2016.csv")
```

```python
# Dando um bizu nos dados para ver se estão ok.
df.head()
```

```python
# Analisando a dimensão da massa de dados.
print(df.shape)
```

```python
# Temos 167 colunas, melhor olhar o dicionário de dados.

# Olhando o dicionário de dados, existem várias informações de natureza qualitativa (que dependeriam de uma análise contextual para serem aproveitadas).

# Considerando o tempo para a análise, dividirei o foco do estudo em três frentes:
# 1-) Estudo da relação entre os dados geográficos dos candidatos e seu desempenho.
# ...
# 2-) Estudo da relação entre alguns dados socio-econômicos e seu desempenho final*.
# * Considero o desempenho final como a somatória das pontuações das provas.

# Vamos para a primeira parte, farei um filtro para selecionar as colunas relevantes para um novo dataframe.
df1 = df.filter(["SG_UF_RESIDENCIA", "NO_MUNICIPIO_RESIDENCIA", "NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO" ])

# Dando uma olhada se filtrei corretamente.
df1.head()
```

```python
# Tem alguns campos NaN que não serão úteis na análise, vou filtrá-los para avaliar somente os candidatos que fizeram o Enem completo.
# Mas eliminarei apenas os candidatos que não fizeram nenhuma das provas, por isso o threshold de 5 no filtro.
df1 = df1.dropna(thresh=5)

# Dando uma olhada se filtrei corretamente.
df1.head()
```

```python
# Filtrando os alunos que não fizeram nenhuma das provas, reduziu-se significativamente a quantidade de alunos.
print(df1.shape)
```

```python
# Realizando um agrupamento por estado.
df1g = df1.groupby("SG_UF_RESIDENCIA")

# Calculando a média e o min/max por componente de avaliação (amplitude).
df1g.aggregate({"NO_MUNICIPIO_RESIDENCIA":['count'], "NU_NOTA_CN":['min', 'mean', 'max'], "NU_NOTA_CH":['min', 'mean', 'max'], "NU_NOTA_LC":['min', 'mean', 'max'], "NU_NOTA_MT":['count', 'min', 'mean', 'max'], "NU_NOTA_REDACAO":['min', 'mean', 'max']}) 
```

```python
# Não ficou muito legível essa avalanche de números, aparentemente as médias estão muito próximas entre os estados, então não dá para destacar facilmente os outliers.
# Vou montar uns gráficos para ver se visualmente fica mais fácil identificar esses outliers, inicialmente vou fazer um catplot com caixas (candles) que identificam a média, o primeiro e último quartis, mas deixam os outliers separados, o que será útil para o nosso cenário.
# Farei um boxplot para cada componente inicialmente, depois farei a análise.

import matplotlib.pyplot as plt
import seaborn as sns

# Componente de Ciências da Natureza.
sns.catplot(x="NU_NOTA_CN", y="SG_UF_RESIDENCIA", kind='box', data=df1)
```

```python
# Componente de Ciências Humanas.
sns.catplot(x="NU_NOTA_CH", y="SG_UF_RESIDENCIA", kind='box', data=df1)
```

```python
# Componente de Linguagens e Códigos.
sns.catplot(x="NU_NOTA_LC", y="SG_UF_RESIDENCIA", kind='box', data=df1)
```

```python
# Componente de Matemática.
sns.catplot(x="NU_NOTA_MT", y="SG_UF_RESIDENCIA", kind='box', data=df1)
```

```python
# Componente de Redação.
sns.catplot(x="NU_NOTA_REDACAO", y="SG_UF_RESIDENCIA", kind='box', data=df1)
```

```python
# Observando as cinco componentes de notas do Enem, percebe-se que médias por estado não se alteram tanto.
# No entanto, existem estados que ficam para trás em todas as componentes em relação aos outros estados, como o Amazonas, Amapá, Piauí, Alagoas e Maranhão, estados estes que também possuem um IDH baixo.
# Olhando na ótica de melhoria do sistema de ensino atual, seriam os estados nas quais se poderia melhorar os esforços pedagógicos e didáticos para evoluir gradualmente os indicadores de desempenho no Enem.

# Vou agora fazer um count por estado:
sns.countplot(y="SG_UF_RESIDENCIA", data=df1)
```

```python
# Existe um viés forte com relação aos dados de SP e do CE por conta da quantidade amostral de dados, lembrando que essa contagem é com base nos alunos que fizeram alguma das provas do Enem.
# Esse viés deve ser levado em conta em análises mais aprofundadas.

# Agora vou avaliar a contagem em outra ótica, considerando os alunos "desistentes", que não fizeram alguma das provas.
df2 = df.filter(["SG_UF_RESIDENCIA", "NO_MUNICIPIO_RESIDENCIA", "NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT", "NU_NOTA_REDACAO" ])
df2 = df2[df2.isnull().any(1)]
sns.countplot(y="SG_UF_RESIDENCIA", data=df2)
```
